```python
import numpy as np
import matplotlib.pyplot as plt
hbar = 1.0
mu = 1.0
e = 1
a=1
n = int(input("Enter the size of the matrix: "))
x1 = -6.0
x2 = 15.0
r = x2 - x1
h = r / n
k=3
b=-.2

K = 2.0 * np.identity(n)
for i in range(0, n - 1):
    K[i, i + 1] = -1
    K[i + 1, i] = -1
    K1=K*(hbar**2/(2*mu*h*h))

my_range = np.linspace(x1, x2, n)
V = np.zeros((n, n))
for i in range(0, n):
    V[i, i] = 0.5*k*(x1+i*h)**2 + 0.5*b*(x1+i*h)**3
print("Plot of Potential")
plt.plot(my_range,np.diag(V))
plt.grid()
plt.show()

A = np.add(K1, V)

w, v = np.linalg.eigh(A)

d=v/np.sqrt(h)
ground_state = d[:, 0]
first_excited_state = d[:, 1]
print("plot of corresponding wavefuctions")
plt.plot(my_range, ground_state)
plt.plot(my_range, first_excited_state)
plt.grid()
plt.show()
```
