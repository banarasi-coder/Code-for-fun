```python
import numpy as np
import matplotlib.pyplot as plt
hbar = 1.0
mu = 1.0
e = 1
a=1
n = int(input("Enter the size of the matrix: "))
x1 = 0.1
x2 = 10.0
r = x2 - x1
h = r / n

K = 2.0 * np.identity(n)
for i in range(0, n - 1):
    K[i, i + 1] = -1
    K[i + 1, i] = -1
    K1=K*(hbar**2/(2*mu*h*h))

my_range = np.linspace(0, r, n)
V = np.zeros((n, n))
for i in range(0, n):
    V[i, i] = (-e**2/(x1+i*h))*np.exp(-(x1+i*h)/a)
print("Plot of Potential")
plt.plot(my_range,np.diag(V))
plt.grid()
plt.show()

A = np.add(K1, V)

w, v = np.linalg.eigh(A)

d=v/np.sqrt(h)
d1=np.zeros([n+2])
d2=np.zeros([n+2])
d3=np.zeros([n+2])
d1[1:n+1]=d[:,0]
d2[1:n+1]=d[:,1]
d3[1:n+1]=d[:,2]
my_range1 = np.linspace(0, r, n+2)
print("plot of corresponding wavefuctions")
plt.plot(my_range1, d1)
plt.plot(my_range1, d2)
plt.grid()
plt.show()
```
