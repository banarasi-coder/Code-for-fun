```python
import numpy as np
import matplotlib.pyplot as plt
import random
p = int(input("Enter the number of coins: "))
n = int(input("Enter the number of times you want to toss the coin: "))
def sample_space(p, n):
    NH = np.zeros(p + 1)
    outcome = np.zeros(p)
    for i in range(n):
        for j in range(p):
            outcome[j] = random.randint(0, 1)
        m = np.sum(outcome)
        NH[int(m)] += 1
    return NH
k = sample_space(p, n)
def create_list(r1, r2):
    return list(map(lambda x: x, range(r1, r2+1)))
y = create_list(1, p + 1)
plt.figure(figsize=(10, 6))
plt.plot(y, k)
plt.title("Histogram")
plt.xlabel("No. of Head")
plt.ylabel("All possible macrostate")
plt.show()
```
