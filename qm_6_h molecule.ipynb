```python
import numpy as np
import matplotlib.pyplot as plt
hbar = 1.0
mu = 1.0
n = int(input("Enter the size of the matrix: "))
x1 = 1
x2 = 10
r = x2 - x1
h = r / n
D=100000
r0=1
a=2
#rp=((x1+i*h)-r0)/(x1+i*h)

K = 2.0 * np.identity(n)
for i in range(0, n - 1):
    K[i, i + 1] = -1
    K[i + 1, i] = -1
    K1=K*(hbar**2/(2*mu*h*h))

my_range = np.linspace(0, r, n)
V = np.zeros((n, n))
for i in range(0, n):
  r = x1 + i * h
  rp = (r - r0) / r
  V[i, i] = D * (np.exp(-2 * a * rp) - np.exp(-a * rp))
print(V)
print("Plot of Potential")
plt.plot(my_range,np.diag(V))
plt.grid()
plt.show()

A = np.add(K1, V)

w, v = np.linalg.eigh(A)

d=v/np.sqrt(h)
ground_state = d[:, 0]
first_excited_state = d[:, 1]
print("plot of corresponding wavefuctions")
plt.plot(my_range, ground_state)
plt.plot(my_range, first_excited_state)
plt.grid()
plt.show()
```
