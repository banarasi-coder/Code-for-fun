```python
import numpy as np
import matplotlib.pyplot as plt
hbar = 1.0
mu = 1.0
e = 1
a=1
n = int(input("Enter the size of the matrix: "))
x1 = 0.1
x2 = 10.0
r = x2 - x1
h = r / n

my_range = np.linspace(0, r, n)
V1 = np.zeros((n, n))
for i in range(0, n):
    V1[i, i] = (-e**2/(x1+i*h))*np.exp(-(x1+i*h)/a)

V = np.zeros((n, n))
for i in range(0, n):
    V[i, i] = -e**2/(x1+i*h)

plt.plot(my_range,np.diag(V1),color='r')
plt.plot(my_range,np.diag(V))
plt.grid()
plt.show()
```
