```python
import numpy as np

n = int(input("Enter the size of the square matrix A: "))

print("Enter the elements of the matrix A (row-wise):")
A = np.zeros((n, n))
for i in range(n):
    A[i] = list(map(float, input().split()))

print("Enter the initial guess for the eigenvector x_0:")
x_0 = np.array(list(map(float, input().split())))

for i in range(100):
    x_k = np.dot(A, x_0)
    x_k = x_k / np.linalg.norm(x_k)
    x_0 = x_k

lambda_max = np.dot(x_k, np.dot(A, x_k)) / np.dot(x_k, x_k)

print("Approximated largest eigenvalue:", lambda_max)
print("Approximated eigenvector:", x_k)
```
