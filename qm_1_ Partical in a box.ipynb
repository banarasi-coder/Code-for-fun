```python
import numpy as np
import matplotlib.pyplot as plt
hbar = 1.0
mu = 1.0
e = 1.0
n = int(input("Enter the size of the matrix: "))
K = 2.0 * np.identity(n)
'''(hbar**2/2*mu)*(2.0 * np.identity(n))'''
for i in range(0, n - 1):
    K[i, i + 1] = -1
    K[i + 1, i] = -1
K1=K*(hbar**2/2*mu)
print(K1)
x1 = -1
x2 = 5
x = x2 - x1
h = x / n

V = np.zeros((n, n))
for i in range(0, n):
    V[i, i] = 0
print(V)
my_range = np.linspace(0, x, n)
plt.plot(my_range, V)
plt.grid()
plt.show()

A = np.add(K, V)
w, v = np.linalg.eigh(A)
print("Eigenvalues of the matrix A:\n", w)
print("Eigenvectors of the matrix A:\n", v)
v1=np.zeros([n+2])
v2=np.zeros([n+2])
v3=np.zeros([n+2])
v1[1:n+1]=v[:,0]
v2[1:n+1]=v[:,1]
v3[1:n+1]=v[:,2]
my_range = np.linspace(0, x, n+2)
plt.plot(my_range,v1)
plt.plot(my_range,v2)
plt.plot(my_range,v3)
plt.grid()
plt.show()
```
