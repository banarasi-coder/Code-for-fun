```python
import numpy as np
import matplotlib.pyplot as plt

hbar = 1.0
mu = 1.0
e = 1.0
n = int(input("Enter the size of the matrix: "))
K = 2.0 * np.identity(n)
'''(hbar**2/2*mu)*(2.0 * np.identity(n))'''
for i in range(0, n - 1):
    K[i, i + 1] = -1
    K[i + 1, i] = -1
    K1=K*(hbar**2/2*mu)
print(K1)

x1 = 0
x2 = 10
x = x2 - x1
h = x / n
k=1
v0=4
c=5


V = np.zeros((n, n))
vdiag=np.zeros(n)
for i in range(0, n):
    r=x1+i*h
    if (r<c):
        V[i, i] =v0
        vdiag[i]=V[i,i]
    else:
        V[i, i] =0
print(V)


my_range = np.linspace(x1, x2, n)
plt.plot(my_range,vdiag)
plt.grid()
plt.show()

A = np.add(K1, V)

w, v = np.linalg.eigh(A)
print("Eigenvalues of the matrix A:\n", w)
print("Eigenvectors of the matrix A:\n", v)

plt.plot(my_range,v[:,0])
plt.plot(my_range,v[:,1])
plt.plot(my_range,v[:,2])
plt.grid()
plt.show()
```
