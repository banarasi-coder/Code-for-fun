```python
3import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
h = 6.6260715*10** -34
k = 1.380649 * 10**-23
N = 6.022140857 * 10**23
f=1*10**13
name = input("Enter the name of Gas: ")
M = float(input(f"Enter the molar mass of {name} (g/mol) :"))
m = M / N * 10**-3
n = int(input("The number of Temperatures: "))
T=[]
for i in range(n):
    l=(i*11)+1
    T.append(l)
CvDP = []
for j in range(n):
    CvDP.append(3*N*k)
plt.plot(T,CvDP,'+',label='Dulong- petits law')
CvE = []
for j in range(n):
    Te = (h*f)/k
    a=Te/T[j]
    b=np.exp(a)/((np.exp(a)-1)**2)
    CvE.append((3*N*k*(a)**2)*(b))
plt.plot(T,CvE,'*',label='Einstein model')
CvD = []
for j in range(n):
    Td = (h*f)/k
    c=(T[j]/Td)**3
    def A(t):
        return (t**4 * np.exp(t)) / (np.exp(t) - 1) ** 2
    result, _ = quad(A, 0,Td/T[j])
    CvD.append(9*N*k*c*result)
plt.plot(T,CvD,label='Debye model')
plt.legend()
plt.title("Cv v/s T")
plt.xlabel("T")
plt.ylabel("Cv")
plt.show()
```
