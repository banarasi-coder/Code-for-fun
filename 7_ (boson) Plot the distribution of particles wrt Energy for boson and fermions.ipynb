```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

S = 1
h = 1
V = 1
m = 1
c = 1

Cn = ((2 * S + 1) * 2 * np.pi * V * (2 * m) ** (3 / 2)) / h ** 3
Cr = ((2 * S) * 4 * V * np.pi) / (h ** 3) * (c ** 3)


# relat BE dist
def rel_boson_dis(e, mu, T):
    return 1 * Cr * (e_range) ** 2 / (np.exp((e - mu) / T) - 1)


# non-relat BE dist
def nonrel_boson_dis(e, mu, T):
    return 1 * Cn * np.sqrt(e_range) / (np.exp((e - mu) / T) - 1)


# relativistic FD dist
def rel_fermi_dis(e, mu, T):
    return 1 * Cr * (e_range) ** 2 / (np.exp((e - mu) / T) + 1)


# non-relat FD dist
def nonrel_fermi_dis(e, mu, T):
    return 1 * Cn * np.sqrt(e_range) / (np.exp((e - mu) / T) + 1)


e_range = np.linspace(0, 10, 1000)

mu = 1.0
T_high = 2.0
T_low = 0.5

fig = plt.figure(figsize=(16, 12))

# Rel bosons
ax1 = fig.add_subplot(221, projection='3d')
ax1.plot(e_range, rel_boson_dis(e_range, mu, T_high), e_range, zdir='y', label='High T')
ax1.plot(e_range, rel_boson_dis(e_range, mu, T_low), e_range, zdir='y', label='Low T')
ax1.set_xlabel('Energy (e)')
ax1.set_ylabel('dN/de')
ax1.set_zlabel('Distribution')
ax1.set_title('Relativistic Bosons')
ax1.legend()

# Non-rel bosons
ax2 = fig.add_subplot(222, projection='3d')
ax2.plot(e_range, nonrel_boson_dis(e_range, mu, T_high), e_range, zdir='y', label='High T')
ax2.plot(e_range, nonrel_boson_dis(e_range, mu, T_low), e_range, zdir='y', label='Low T')
ax2.set_xlabel('Energy (e)')
ax2.set_ylabel('dN/de')
ax2.set_zlabel('Distribution')
ax2.set_title('Non-relativistic Bosons')
ax2.legend()

# Rel fermions
ax3 = fig.add_subplot(223, projection='3d')
ax3.plot(e_range, rel_fermi_dis(e_range, mu, T_high), e_range, zdir='y', label='High T')
ax3.plot(e_range, rel_fermi_dis(e_range, mu, T_low), e_range, zdir='y', label='Low T')
ax3.set_xlabel('Energy (e)')
ax3.set_ylabel('dN/de')
ax3.set_zlabel('Distribution')
ax3.set_title('Relativistic Fermions')
ax3.legend()

# Non-rel fermions
ax4 = fig.add_subplot(224, projection='3d')
ax4.plot(e_range, nonrel_fermi_dis(e_range, mu, T_high), e_range, zdir='y', label='High T')
ax4.plot(e_range, nonrel_fermi_dis(e_range, mu, T_low), e_range, zdir='y', label='Low T')
ax4.set_xlabel('Energy (e)')
ax4.set_ylabel('dN/de')
ax4.set_zlabel('Distribution')
ax4.set_title('Non-relativistic Fermions')
ax4.legend()

plt.tight_layout()
plt.show()
```
