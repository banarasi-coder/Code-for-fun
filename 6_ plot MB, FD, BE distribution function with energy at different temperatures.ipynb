```python
import numpy as np
import matplotlib.pyplot as plt
k = 0.01
mu = -1
n = int(input("Enter the number of Temperatures and Energy points: "))
E = []
T = []
for i in range(n):
    l = i * 1 - 1000
    z = i * 1 + 1
    E.append(l)
    T.append(z)
Fmb = []
for j in range(n):
    a = np.exp((E[j] - mu) / (k * T[j]))
    Fmb.append(1 / a)
plt.subplot(2, 2, 1)
plt.plot(E, Fmb,  label='Maxwell Boltzmann')
plt.ylim(0,10)
plt.xlim(-100,100)
Ffd = []
for j in range(n):
    b = np.exp((E[j] - mu) / (k * T[j]))
    Ffd.append(1 / (b + 1))
plt.subplot(2, 2, 2)
plt.plot(E, Ffd, label='Fermi-Dirac')
plt.ylim(0,10)
plt.xlim(-100,100)
Fbe = []
for j in range(n):
    c = np.exp((E[j] - mu) / (k * T[j]))
    Fbe.append(1 / (c - 1))
plt.subplot(2, 2, 3)
plt.plot(E, Fbe, label='Bose-Einstein')
plt.ylim(0,10)
plt.xlim(-100,100)

plt.subplot(2, 2, 4)
plt.plot(E,Fbe,label='Bose Einstein')
plt.plot(E,Ffd,'+',label='Fermi- Didac')
plt.plot(E,Fmb,'*',label='Maxwell Boltzmann')
plt.ylim(0,10)
plt.xlim(-100,100)
plt.xlabel("Energy (E)")
plt.ylabel("Distribution Function (f(E))")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

k = 1
n = int(input("The number of Temperatures and Energy: "))
E = []
T = []
mu_values = [-1, 0, 1]

fig, axes = plt.subplots(nrows=len(mu_values), ncols=3, figsize=(15, 5*len(mu_values)))

for i in range(n):
    l = i + 1
    E.append(l)
    T.append(l)

for row, mu in enumerate(mu_values):
    Fmb = []
    for j in range(n):
        a = np.exp((E[j] - mu) / (k * T[j]))
        Fmb.append(1 / a)
    axes[row, 0].plot(T, Fmb)
    axes[row, 0].set_title(f'mu={mu} - Fmb')

    Ffd = []
    for j in range(n):
        b = np.exp((E[j] - mu) / (k * T[j]))
        Ffd.append(1 / (b + 1))
    axes[row, 1].plot(T, Ffd)
    axes[row, 1].set_title(f'mu={mu} - Ffd')

    Fbe = []
    for j in range(n):
        c = np.exp((E[j] - mu) / (k * T[j]))
        Fbe.append(1 / (c - 1))
    axes[row, 2].plot(T, Fbe)
    axes[row, 2].set_title(f'mu={mu} - Fbe')

for ax in axes.flat:
    ax.set(xlabel='T', ylabel='f(E)')
    ax.label_outer()

plt.tight_layout()
plt.show()
```
