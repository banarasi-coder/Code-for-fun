```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la

print("Matrix")
n=int(input("Enter the size of a matrix"))
A=np.random.rand(n,n) * 1j + np.random.rand(n,n)
print(A)

print("Complex cojugate of matrix")
B=np.conj(A).T
print(B)

print("Sum of randum matrix and it's transpose")
C=np.add(A,B)
print(C)

w, v = np.linalg.eig(C)
print("Eigen values:\n",w)
print("eigenvectors:\n",v)

print("Complex cojugate of eigen vector")
D=np.conj(v).T
print(D)
print("Multiplication of Eigen vector and its hermitian conjugate")
np.matmul(D,v)
```
