```python
# Consider a system of 3 energy level and 2 particles for MB/BE/FD
import numpy as np
import matplotlib.pyplot as plt
k = 1
E1 = 1
E2 = 2
E3 = 3

def p_f_MB(beta):
    return np.exp(-beta * 2*E1) + np.exp(-beta * 2*E2) + np.exp(-beta * 2*E3)+ 2*(np.exp(beta * (-E1-E2))) + 2*(np.exp(beta * (-E2-E3))) + 2*(np.exp(beta * (-E3-E1)))

def p_f_BE(beta):
    return (np.exp(-beta * 2*E1)) +(np.exp(-beta * 2*E2)) + (np.exp(-beta * 2*E3))+ (np.exp(beta * (-E1-E2))) + (np.exp(beta * (-E2-E3))) + (np.exp(beta * (-E3-E1)))


def p_f_FD(beta):
    return (np.exp(beta * (-E1-E2))) + (np.exp(beta * (-E2-E3))) + (np.exp(beta * (-E3-E1)))

temp_range = np.linspace(0.1, 5, 100)
beta = 1 / (k*temp_range)

Z_MB = p_f_MB(beta)
Z_BE = p_f_BE(beta)
Z_FD = p_f_FD(beta)

plt.figure(figsize=(10, 6))
plt.plot(temp_range, Z_MB,'--', label='Maxwell-Boltzmann')
plt.plot(temp_range, Z_BE,'*', label='Bose-Einstein')
plt.plot(temp_range, Z_FD, label='Fermi-Dirac')

plt.xlabel('Temperature')
plt.ylabel('Partition Function (Z)')
plt.title('Partition Function vs Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

k = 8.617e-5  # Boltzmann constant in eV (Kb/e)
E = [0.00, 0.01]  # Energy in eV
T = np.arange(0, 501, 5)  # Temperature range in Kelvin

# Partition function
def f(T):
    return (np.exp(-E[0] / (k * T)) + np.exp(-E[1] / (k * T)))**2

# Log of partition function
def partition(T):
    return np.log((np.exp(-E[0] / (k * T)) + np.exp(-E[1] / (k * T)))**2)

# Internal Energy
def inter(T):
    return k * T**2 * np.gradient(partition(T), T)

# Specific heat capacity
def cv(T):
    return np.gradient(inter(T), T) / k

# Helmholtz function
def f4(T):
    return -k * T * partition(T)

# Entropy
def f5(T):
    return -np.gradient(f4(T), T) / k

# Plotting the functions
plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.plot(T, f(T), "-k")
plt.title("Partition function vs Temperature", fontsize=14)
plt.xlabel("T (Kelvin)", fontsize=12)
plt.ylabel("Z", fontsize=12)
plt.grid()

plt.subplot(2,2,2)
plt.plot(T, inter(T), "-k")
plt.title("Internal Energy vs Temperature", fontsize=14)
plt.xlabel("T (Kelvin)", fontsize=12)
plt.ylabel("U(eV)", fontsize=12)
plt.grid()

plt.subplot(2,2,3)
plt.plot(T, cv(T), "-k")
plt.title("Molar Heat capacity vs Temperature", fontsize=14)
plt.xlabel("T (Kelvin)", fontsize=12)
plt.ylabel("Cv/k (eV/K)", fontsize=12)
plt.grid()

plt.subplot(2,2,4)
plt.plot(T, f5(T), "-k")
plt.title("Entropy vs Temperature", fontsize=14)
plt.xlabel("T(Kelvin)", fontsize=12)
plt.ylabel("S/k(eV/K)", fontsize=12)
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 8.617333262145e-5  # Boltzmann constant in eV/K
E = [0.0, 0.01]        # Energy levels in eV

# Probability functions
def p1(T):
    return 1 / (1 + np.exp(-E[1] / (k * T)))  # for single particle in E1

def p2(T):
    return np.exp(-E[1] / (k * T)) / (1 + np.exp(-E[1] / (k * T)))  # for single particle in E2

# Temperature range
T = np.linspace(1, 1000, 1000)  # Temperature range from 1 K to 1000 K

# Plotting
plt.plot(T, p1(T), "*-k", label="For E1=0.00 eV")
plt.plot(T, p2(T), "-", label="For E2=0.01 eV")
plt.title("Probability of Occupancy vs Temperature in Energy level", fontsize=14)
plt.xlabel("T (Kelvin)", fontsize=12)
plt.ylabel("Probability of Occupancy", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
