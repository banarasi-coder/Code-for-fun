```python
import numpy as np
import matplotlib.pyplot as plt

h = 6.626e-34
k = 1.38e-23
c = 3e8
t = np.array([1000, 1200, 1400])
n = 1400
r = np.linspace(0.1e-6, 15e-6, n)
r0 = np.linspace(5e-6, 15e-6, n)

u = np.zeros((3, n))
m = np.zeros((3, n))
v = np.zeros((3, n))

for i in range(n):
    for j in range(3):
        u[j, i] = (8 * np.pi * h * c) / ((r[i]) ** 5 * (np.exp(h * c / (r[i] * k * t[j])) - 1))
        m[j, i] = (8 * np.pi * k * t[j]) / ((r0[i]) ** 4)
        v[j, i] = (8 * np.pi * h * c) / ((r[i]) ** 5 * (np.exp(h * c / (r[i] * k * t[j]))))

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
for j in range(3):
    plt.plot(r, u[j, :], label=f"T={t[j]}")
plt.xlabel("Wavelength (m)")
plt.ylabel("Energy")
plt.title("Planck's Law")
plt.legend()
plt.grid()

plt.subplot(2, 2, 2)
for j in range(3):
    plt.plot(r0, m[j, :], label=f"T={t[j]}")
plt.xlabel("Wavelength (m)")
plt.ylabel("Energy")
plt.title("Rayleigh Law")
plt.legend()
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(r, u[2, :], 'b*', label="Planks")
plt.plot(r0, m[2, :], 'g', label="Rayleigh")
plt.xlabel("Wavelength (m)")
plt.ylabel("Energy")
plt.title("Comparison between Planck's and Rayleigh Law")
plt.legend()
plt.text(1.2e-05, 500, "T=1200 K")
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(r, u[2, :], 'r', label="Planck")
plt.plot(r0, m[2, :], 'g*', label="Rayleigh")
plt.plot(r, v[2, :], 'b--', label="Weins")
plt.xlabel("Wavelength (m)")
plt.ylabel("Energy")
plt.title("Comparison between Planck's, Rayleigh, and Weins Law")
plt.legend()
plt.text(1.2e-05, 450, "T=1200 K")
plt.grid()

plt.tight_layout()
plt.show()
```
