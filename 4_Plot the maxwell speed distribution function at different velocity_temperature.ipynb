```python
import numpy as np
import matplotlib.pyplot as plt

k = 1.380649 * 10**-23
N = 6.022140857 * 10**23

b = input("Enter the name of Gas: ")
M = float(input(f"Enter the molar mass of {b} (g/mol) :"))

n = 3
v = np.linspace(0, 2000, 1000)

T_values = [300, 500, 700]

m = M / N * 10**-3

plt.figure(figsize=(10, 6))

for T in T_values:
    Vmp = np.sqrt((2 * k * T) / m)
    Vav = np.sqrt(4 / (np.pi)) * Vmp
    Vrms = np.sqrt(3/2) * Vmp
    a = m / (2 * k * T)
    f = 4*np.pi*(a / np.pi)**(3/2) * v**2 * np.exp(-a*(v**2))
    plt.plot(v, f,"*", label=f'Temperature = {T} K')

    print(f"For Temperature = {T} K:")
    print("The values of Vmp:", Vmp)
    print("The values of Vav:", Vav)
    print("The values of Vrms:", Vrms)

plt.title("f(v) vs v")
plt.xlabel("v (m/s)")
plt.ylabel("f(v)")
plt.legend()
plt.show()
```
